```python
FILE_NAME = "btc_sample.xlsx"
data = pd.read_excel(FILE_NAME) #Read data
data
```

```python
data.isnull().sum() #Check the null status
```

```python
data.describe() #View Data Distribution
```

```python
df = data[['time_close', 'quote.USD.close', 'quote.USD.open', 'quote.USD.high', 'quote.USD.low']] #Select partial data
df
```

```python
def time_handle(x):
    return x.replace('T', ' ').split(' ')[0] 
#Process the time and select the number of days(The effective feature values cannot be extracted by daily data alone)

df['time_close'] = df['time_close'].map(time_handle)
```

```python
df
```

```python
df['quote.USD.close'].plot()
```

```python
df.columns
```

```python
df['quote.USD.open'].plot()
```

```python
id_list = []
#Put the data of every seven days together to facilitate the extraction of features

for index in df.index:
    id_list.append((index +1 )// 7)

id_list
```

```python
df['id'] = id_list #Use time as ID
```

```python
df = df.sort_values(by=['time_close'])
#Sorting by time_close
```

```python
df = df.drop(['time_close'], axis=1)
```

```python
df = df.reset_index()
df = df.rename(columns={'index': 'time'})
```

```python
df.dtypes
```

```python
df.head()
```

```python
from tsfresh.utilities.dataframe_functions import impute
extracted_features = extract_features(df, column_sort="time",  column_id="id", impute_function=impute) 
#Extract multiple features
```

```python
extracted_features = extracted_features.reset_index() #Reset index
```

```python
extracted_features.head() 
```

```python
extracted_features.shape #View data feature dimensions
```

```python
extracted_features.columns 
```

```python
extracted_features.head()
```

```python
new_data = []
index_list = []
for temp in df.groupby('id')['quote.USD.close']:
    if len(temp[1].values) == 7:
        index_list.append(temp[0])
        new_data.append(temp[1].values)
#data reorganization for closing data
```

```python
new_df = pd.DataFrame(new_data, columns=[f'day{i+1}' for i in range(7)]) #Building new data
```

```python
new_df.shape
```

```python
new_df['id'] = index_list
```

```python
new_df = pd.merge(new_df, extracted_features, left_on='id', right_on='index', how='left') # 拼接数据
new_df = new_df.drop(['id', 'index'], axis=1)
```

```python
new_df.head()
```

```python
#Building Inputs and Outputs
x = new_df.loc[:, ~(new_df.columns.isin([f'day{i+1}' for i in range(7)]))]
y = new_df.loc[:, new_df.columns.isin([f'day{i+1}' for i in range(7)])].to_numpy()
```

```python
#Feature selection
#May proceed very slowly

from tsfresh import select_features
target_columns = []
for i in range(7):
    temp = select_features(x, y[:, i])
    target_columns.extend(temp.columns.to_list())

target_columns = list(set(target_columns))
```

```python
len(target_columns)
```

```python
new_x = x.loc[:, target_columns]
```

```python
new_x.isnull().sum().sum()
new_x.head()
```

```python
y
```

```python
#Dividing the training set and test set
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(new_x, y, test_size=.2, random_state=123)

```

```python
from sklearn.ensemble import RandomForestRegressor

#Building the model
model = RandomForestRegressor(random_state=123)
#Training
model.fit(train_x, train_y)

```

```python
y_pred = model.predict(test_x) #Prediction
```

```python
y_pred[:, 1]
```

```python
#Calculating evaluation metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mae = mean_absolute_error(test_y, y_pred)
mse = mean_squared_error(test_y, y_pred)
r2 = r2_score(test_y, y_pred)

print(mae, mse, r2)  #From R2, the fit coefficient is close to 1 and the model is good.
```

```python
import matplotlib.pyplot as plt
```

```python
#Predicting
for i in range(7):
    fig = plt.figure(figsize=(15, 10))
    plt.title(f"Day{i+1}")
    plt.plot(range(len(y_pred)), test_y[:, i], label='True Value')
    plt.plot(range(len(y_pred)), y_pred[:, i], label="Predicted Value")
    plt.legend()
    plt.show()
    
#Predictions for the coming week in groups of seven days
```
